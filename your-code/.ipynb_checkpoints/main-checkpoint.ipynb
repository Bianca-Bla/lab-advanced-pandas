Now create the standardized columns `CGPA_std`, `GRE_std`, and `LOR_std` and add them to the `admissions` dataframe.

```python
admissions = pd.DataFrame(columns = ['CGPA_std', 'GRE_std', 'LOR_std'] )
```

```python
# your code here
admissions['CGPA_std'] = (admissions_predict['CGPA'] - admissions_predict['CGPA'].mean()) / admissions_predict['CGPA'].std()
admissions['GRE_std'] = (admissions_predict['GRE Score'] - admissions_predict['GRE Score'].mean()) / admissions_predict['GRE Score'].std()
admissions['LOR_std'] = (admissions_predict['LOR'] - admissions_predict['LOR'].mean()) / admissions_predict['LOR'].std()
admissions
```

We will generate the decision choice at random using the code below. Please run the cell.

```python
# Libraries
from random import choices
```

```python
std_columns = ['CGPA_std', 'GRE_std', 'LOR_std']

decision_choice = choices(std_columns, k=admissions.shape[0])
 
```

```python
# your code here
admissions['deciding_column'] = admissions.lookup(admissions.index,decision_choice)
admissions.head()
```

Create a column called `decision` in the `admissions` dataframe. Assign 1 to this column if the value of `deciding_column` is greater than 0.8 and 0 otherwise.

```python
# your code here
admissions['decision'] = np.where(admissions['deciding_column'] > 0.8, 1, 0)
admissions.sample(5)
```

How many applicants will be accepted to the program using the decision column? Compute the result below.

```python
# your code here --> 87 candidates will be accepted
admissions['decision'].value_counts()
```

In the cell below, give all student with a university rating of 4 or higher a 10 point boost on their GRE score and split the column into 4 bins using the `cut` function. Assign this new score to the variable `adjusted_gre`.

```python
# your code here

admissions_predict['adjusted_gre'] = ''
```

```python
for row in admissions_predict['adjusted_gre']:
    if [admissions_predict['University Rating'] >= 4]:
        admissions_predict['adjusted_gre'] = admissions_predict['GRE Score'] + 10
    else:
        admissions_predict['adjusted_gre'] = ''
```

```python
admissions_predict.sample(10)
```

```python


display(admissions_predict['adjusted_gre'].sort_values())
```

```python
bin_labels = ['Low', 'Medium', 'Mid-High', 'High']

admissions_predict['bins'] = pd.cut(admissions_predict['adjusted_gre'], 4, labels = bin_labels)

admissions_predict.sample(10)
```
